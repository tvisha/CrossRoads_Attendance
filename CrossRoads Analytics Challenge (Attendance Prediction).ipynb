```python
def acct1(num):
    if num==1:
        return '1'
    elif num==2:
        return '2'
    elif num==3:
        return '3'
    
    elif num==4:
        return '4'
    
    elif num==5:
        return '5'
    
    elif num==6 & num<=10:
        return '6-10'
    elif num==11 & num<=50:
        return '11-50'
    else:
        return '>51'
    
```

```python
df2['acct_type_desc'] = df2.groupby('event_name')['acct_type_desc'].apply(lambda x:x.fillna(x.value_counts().idxmax()))
df2['plan_event_name'] = df2.groupby('event_name')['plan_event_name'].apply(lambda x:x.fillna(x.value_counts().idxmax()))
df2['comp_name'] = df2.groupby('event_name')['comp_name'].apply(lambda x:x.fillna(x.value_counts().idxmax()))
df2['Tenure'] = df2.groupby('event_name')['Tenure'].apply(lambda x:x.fillna(x.value_counts().idxmax()))
df2['paid'] = df2.groupby('event_name')['paid'].apply(lambda x:x.fillna(x.value_counts().idxmax()))
```

```python
#Replacing nulls in isSTM column with 0 as it is a flag
df2['isSTM']=df2['isSTM'].fillna(0)
```

```python
#Replacing 1.0 and Yes to 1 and nulls to 0 in Resold column too as it is flag
print(df2['Resold'].value_counts())
resold_mapping={1.0:1,'Yes':1}
df2['Resold'] = df2['Resold'].map(resold_mapping)
df2['Resold']=df2['Resold'].fillna(0)
print(df2['Resold'].value_counts())
#Same with isAttended
df2['isAttended'].replace(['Yes', 'No'],
                        [1, 0], inplace=True)
print(df2['isAttended'].value_counts())
```
